```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')    ## I don't like pandas setting with copy warnings 

df = pd.read_excel(r'..\raw_data\CO2 combination.xlsx','Sheet1')
cf = pd.read_excel(r'task1\All.xlsx','Sheet1')
```

```python
combo1 = df.groupby('Combination1')['Food'].unique()
combo2 = df.groupby('Combination 2')['Food'].unique()
combo1 = combo1.to_frame()
combo2 = combo2.to_frame()
```

```python
for idx, row in combo1.iterrows():  # Iterate through combo1 rows using index and row

    temp = []
    for j in cf.itertuples():  
        if j.Food in row[0]:
            temp.append(j.CO2) 
    if temp:  
        mean = np.mean(temp)
        median = np.median(temp)
        no = len(temp) 
        combo1.at[idx, 'mean_co2'] = mean
        combo1.at[idx, 'median_co2'] = median
        combo1.at[idx, 'no'] = no  
```

```python
for idx, row in combo2.iterrows():  
    temp = []
    for j in cf.itertuples():  
        if j.Food in row[0]:
            temp.append(j.CO2) 
    if temp:  
        mean = np.mean(temp)
        no = len(temp) 
        median = np.median(temp)
        combo2.at[idx, 'mean_co2'] = mean
        combo2.at[idx, 'median_co2'] = median
        combo2.at[idx, 'no'] = no  
```
