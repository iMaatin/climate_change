```python
emro = ['AFG', 'ARE', 'BHR', 'DJI','EGY','IRN','IRQ','JOR',
    'KWT', 'LBN', 'LBR', 'MAR', 'OMN', 'PAK','PSE' , 'QAT','SAU',
    'SDN','SOM', 'SYR', 'YEM','TUN']
drop_list = ['superregion2','age','urban','edu', 'upperci_95',
          'lowerci_95', 'serving', 's_lowerci_95', 's_upperci_95']

def select(col, val,df):
    df = df[df[col] == val]
    return df

### old emro select 
# def emro_select(df:pd.DataFrame):
    
#     d1 = select('age',999,df)
#     d2 = select('edu',999,d1)
#     d3 = select('urban',999,d2)
# #     d3 = d3.drop(drop_list,axis=1)
#     d4 = d3[d3['iso3'].isin(emro)]          ### comment out if u want to calculate for all countries... also replace d4 with d3 in the next 3 lines 

#     all = select('female',999,d4)
#     males = select('female',0,d4)
#     females = select('female',1,d4)
    
#     return all, males, females


### new emro select for adding entries for yearly means
def emro_select(df: pd.DataFrame):
    
    def add_yearly_means(df):
        # Calculate yearly means (excluding any non-numeric or categorical columns like 'female' and 'iso3')
        grouped = df.groupby('year').mean(numeric_only=True).reset_index()
        grouped['iso3'] = 'EMR'
        return pd.concat([df, grouped], ignore_index=True)
    
    d1 = select('age', 999, df)
    d2 = select('edu', 999, d1)
    d3 = select('urban', 999, d2)
    d4 = d3[d3['iso3'].isin(emro)]  # Change to d3 if using all countries


    all = select('female', 999, d4)
    males = select('female', 0, d4)
    females = select('female', 1, d4)

    all = add_yearly_means(all)
    males = add_yearly_means(males)
    females = add_yearly_means(females)

    return all, males, females

### tests below 
# test_path = cf.iloc[1]['path']
# med = cf.iloc[1]['median_co2']
# data = pd.read_csv(test_path)

# a1, m1, f1 = emro_select(data)
# a1.columns

# a1['emissions'] = a1['median'].apply(lambda x: (x /1000)  * med)
# a1
```

```python
def calc_co2(med:float, data:pd.DataFrame):
    data['emissions'] = data['median'].apply(lambda x: (x / 1000)  * med)
    return data[['iso3','year','median','emissions']].reset_index()

for _,item in cf.iterrows():
    item_co2 = item['median_co2']
    path = item['path']
    name = item['name']
    print(f'>>> {name} done!')
    data = pd.read_csv(path)
    all, males, females = emro_select(df=data)

    all2 = calc_co2(med=item_co2,data=all)
    males2 = calc_co2(med=item_co2, data=males)
    females2 = calc_co2(med=item_co2, data=females)

    os.makedirs('emissions_v2',exist_ok=True)
    all2.to_csv(f'emissions_v2\{name}_all.csv',index=False)
    males2.to_csv(f'emissions_v2\{name}_males.csv',index=False)
    females2.to_csv(f'emissions_v2\{name}_females.csv',index=False)
```

```python
from collections import defaultdict

def calc_total (path:str, suff):
    all_glob = glob.glob(f'{path}\*_{suff}.csv')

    keep = []
    
    for i in all_glob:
        temp = pd.read_csv(i)
        for j in zip(temp['iso3'], temp['year'], temp['emissions']):
            keep.append(j)
    
    keep_sum = defaultdict(float)
    for iso3, year, emissions in keep:
        keep_sum[(iso3, year)] += emissions

    result = [(key[0], key[1], total_emissions) for key, total_emissions in keep_sum.items()]
    results_df = pd.DataFrame(result,columns=['Country', 'year', 'emissions'])
    return results_df
```

```python
folder = 'emissions_v2'
for i in ['all', 'males', 'females']:
    temp_df = calc_total(folder, i)
    temp_df.to_csv(f'{folder}\\total_{i}.csv',index=False)
```
