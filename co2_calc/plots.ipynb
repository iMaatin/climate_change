```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import os 

all = pd.read_csv(r'emissions\total_all.csv')
```

```python
sns.set_style("whitegrid")

unique_years = all['year'].unique()

# Loop through each year and generate a scatterplot
for year in unique_years:
    df_year = all[all['year'] == year]
    plt.figure(figsize=(12, 6))  
    sns.scatterplot(data=df_year, y='emissions', x='Country', s=100, alpha=0.7, edgecolor='k')
    plt.xticks(rotation=45, ha='right')  
    plt.title(f'Emissions by Country in {year}', fontsize=16, fontweight='bold') 
    plt.xlabel('Country', fontsize=12)  
    plt.ylabel('Emissions', fontsize=12)
    plt.tight_layout()  
    plt.savefig(f'emissions/plots/all_{year}.png')
    plt.show()
```

```python

def read_n_plot (path:str, name:str, save:bool):
    df = pd.read_csv(path)
    sns.set_style("whitegrid")

    unique_years = df['year'].unique()

    # Loop through each year and generate a scatterplot
    for year in unique_years:
        df_year = df[df['year'] == year]
        plt.figure(figsize=(12, 6))  
        sns.scatterplot(data=df_year, y='emissions', x='Country', s=100, alpha=0.7, edgecolor='k')
        plt.xticks(rotation=45, ha='right')  
        plt.title(f'Emissions by Country in {year}', fontsize=16, fontweight='bold') 
        plt.xlabel('Country', fontsize=12)  
        plt.ylabel('Emissions', fontsize=12)
        plt.tight_layout()
        if save:
            plt.savefig(f'emissions/plots/{name}_{year}.png')
        plt.show()
```

```python
read_n_plot(r'emissions\total_females.csv', 'females', save=True)
```

```python
read_n_plot(r'emissions\total_males.csv', 'males', save=True)
```
