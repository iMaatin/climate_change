```python
import numpy as np 
import pandas as pd 
import os 
import glob
import warnings
warnings.filterwarnings('ignore')    ## I don't like pandas setting with copy warnings 
```

```python
cf = pd.read_csv(r'task2\combo2_v3.csv')
cf.dropna(inplace=True)
cf
```

```python
cf['path'] = cf['gdd_2'].apply(lambda x: f'..\\raw_data\\Global estimates\\{x}_global.csv' )
cf
```

```python
def select(col, val,df):
        df = df[df[col] == val]
        return df

def emro_select(df:pd.DataFrame):
    
    d1 = select('age',999,df)
    d2 = select('edu',999,d1)
    d3 = select('urban',999,d2)

    all = select('female',999,d3)
    males = select('female',0,d3)
    females = select('female',1,d3)
        
    return all, males, females 
```

```python
def calc_co2(med:float, data:pd.DataFrame):
    data['emissions'] = data['median'].apply(lambda x: (x / 1000)  * med)
    return data[['year','median','emissions']].reset_index()

for _,item in cf.iterrows():
    item_co2 = item['median_co2']
    path = item['path']
    name = item['name']
    print(f'>>> {name} done!')
    data = pd.read_csv(path)
    all, males, females = emro_select(df=data)

    all2 = calc_co2(med=item_co2,data=all)
    males2 = calc_co2(med=item_co2, data=males)
    females2 = calc_co2(med=item_co2, data=females)

    os.makedirs('global_emissions',exist_ok=True)
    all2.to_csv(f'global_emissions\{name}_all.csv',index=False)
    males2.to_csv(f'global_emissions\{name}_males.csv',index=False)
    females2.to_csv(f'global_emissions\{name}_females.csv',index=False)
```

```python
from collections import defaultdict

def calc_total (path:str, suff):
    all_glob = glob.glob(f'{path}\*_{suff}.csv')

    keep = []
    
    for i in all_glob:
        temp = pd.read_csv(i)
        for j in zip(temp['year'], temp['emissions']):
            keep.append(j)
    
    keep_sum = defaultdict(float)
    for year, emissions in keep:
        keep_sum[(year)] += emissions

    result = [(key, total_emissions) for key, total_emissions in keep_sum.items()]
    results_df = pd.DataFrame(result,columns=['year', 'emissions'])
    return results_df
```

```python
folder = 'global_emissions'
for i in ['all', 'males', 'females']:
    temp_df = calc_total(folder, i)
    temp_df.to_csv(f'{folder}\\total_{i}.csv',index=False)
```
